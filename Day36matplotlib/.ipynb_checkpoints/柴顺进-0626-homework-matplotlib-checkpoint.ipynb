### 第一步：导入相应的模块

```python
from matplotlib import pyplot as plt
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

### 第二步：用pandas加载航班数据flights.csv

```python
df = pd.read_csv('flights.csv',skiprows=[0],names=['idx','year','month','passengers'])
df.info()
```

### 第三步：用折线图画出乘客数量随年份变化的走势图

```python
# 1.通过年分组，计算出每年对应的乘客数量
s1 = df['passengers'].groupby(df['year']).sum()
years = list(s1.index)
qualities = list(s1)
graph = plt.plot(years,qualities,label='乘客年份变化图')
plt.legend(loc='upper left')
plt.title('乘客数量变化图')
plt.xlabel('年份')
plt.ylabel('乘客量(单位：万)')
plt.grid(True)
plt.ylim(0,6000)
t = 1960
plt.annotate(r'$5714$',
xy=(1960, 5714), xycoords='data',
xytext=(-35, -95), textcoords='offset points', fontsize=16,
arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0"))
```

### 第四步：分析乘客在一年中各月份的分布（柱状图）

```python
year_mon_qualities = df[['passengers']].groupby([df['year'],df['month']]).sum()
year_mon_qualities.plot(kind='bar')
```

```python
for i in range(12):  
    year_mon_qualities[i*12:(i+1)*12].plot(kind='bar')  # matplotlib可以直接把数据索引转换为数据集
```

### 作业二、用matplotlib画出如图2.png数学函数在0到2区间内的曲线图
$$
    f(x) = sin^2(x-2)e^{-x^2}
$$

```python

x=np.linspace(-2-np.pi,2+np.pi,1000)
y = np.sin(x-2)**2*np.exp(-x**2)

# 根据函数绘制出图像
plt.plot(x,y,'b',lw=2.5,label='曲线')
```
