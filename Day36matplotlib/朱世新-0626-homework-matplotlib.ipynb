### 第一步：导入相应的模块

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 第二步：用pandas加载航班数据flights.csv

```python
df = pd.read_csv("flights.csv")
df.head()
```

```python
df['year'].value_counts()
```

### 第三步：用折线图画出乘客数量虽年份变化的走势图

```python
plt.rcParams["font.sans-serif"]=['SimHei']  # 用于正常显示中文标签
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号

pass_nums=df['passengers'].groupby(df['year']).sum()
pass_nums
x_data=list(pass_nums.index)
y_data=list(pass_nums.values)
plt.plot(x_data,y_data,label='乘客数量随年份变化走势图')
plt.xlabel('年份')
plt.ylabel('乘客数量')
plt.legend()
plt.show()
```

### 第四步：分析乘客在一年中各月份的分布（柱状图）

```python
pass_year_mon_passnum = df[['passengers']].groupby([df.year,df.month]).sum()
pass_year_mon_passnum.head()
```

```python
pass_year_mon_passnum.plot(kind='bar')
```
