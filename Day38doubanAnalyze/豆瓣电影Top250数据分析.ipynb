```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
```

```python
# 加载数据
df_1 = pd.read_csv('csv/top250_f1.csv',sep='#')
# 查看加载的数据 前三行
df_1.head(3)
```

```python
# 加载第二个数据
df_2 = pd.read_csv('csv/top250_f2.csv',sep='#')
df_2.head(3)
```

```python
# 根据需要分析的问题，从两个数据集中提取部分信息
df1_cut = df_1[['num','title','init_year','area','genre','rating_num','comment_num']]
df2_cut = df_2[['num','language','director','cast','movie_duration','tags']]
df = pd.merge(df1_cut,df2_cut,on='num',how='outer')
df.head()
```

```python
# 对数据清洗和预处理
# 一、去除genre，language，director，movie_duration，cast
df['genre'] = df['genre'].str[2:-2]
df['language'] = df['language'].str[2:-2]
df['director'] = df['director'].str[2:-2]
df['cast'] = df['cast'].str[2:-2]
df['movie_duration'] = df['movie_duration'].str[2:-2]

df.head()
```
