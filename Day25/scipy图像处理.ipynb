```python
from scipy import misc
from matplotlib import pyplot as plt
from scipy import ndimage
import numpy as np
```

```python
image = misc.ascent()
plt.subplot(221)
img = plt.imshow(image)
plt.axis("off")
```

```python
# 使用scipy的中值滤波处理图片
fislter = ndimage.median_filter(image,size=50)
```

```python
plt.imshow(fislter)
```

```python
# 旋转
rotate = ndimage.rotate(image,90)
plt.imshow(rotate)
```

```python
# 边缘检测
prewitt = ndimage.prewitt(image)
plt.imshow(prewitt)
```

```python
plt.subplot(221) # plt是一个画板，2x2画板，1第一个图片
plt.title("orign")
plt.imshow(image)
plt.subplot(222) # 2x2画板中的第二个
plt.title("median")
plt.imshow(fislter)
plt.subplot(223) # 2x2画板中的第二个
plt.title("rotate")
plt.imshow(rotate)
plt.subplot(224) # 2x2画板中的第二个
plt.title("prewitt")
plt.imshow(prewitt)
```
