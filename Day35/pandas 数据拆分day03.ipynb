```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
```

```python
data = np.random.randn(1000) # 服从正太分布
```

```python
# cut(x, bins, right=True, labels=None, retbins=False, precision=3,include_lowest=False)
```

```python
result = pd.cut(data,4)   # data中的（最大值-最小值）/4（分割数） = 分割数
result
pd.value_counts(result)
```

```python
result = pd.cut(data,4,precision=2)   # precision 保留有效小数点位数
result   
pd.value_counts(result)
```

```python
#qcut(x, q, labels=None, retbins=False, precision=3, duplicates='raise'):
result = pd.qcut(data,4) # 将数据均分成四组，先对数据进行排序，然后再对数据均分为四组
result
pd.value_counts(result)
```

```python
result = pd.qcut(data,[0,0.1,0.5,0.9,1])  # [0,0.1,0.5,0,9,1] 这个数据是0.1 - 0 ；0.5-0.1；0.9-0.5；1-0.9 这个数据当做分位数；还是先排序后分组
pd.value_counts(result)
```

```python
# 如果分位数的差值和小于1的情况下，不会取的全数据 :取完分位数差值所占百分比就不在取得数据
result = pd.qcut(data,[0,0.1,0.3,0.75])
pd.value_counts(result)
```
