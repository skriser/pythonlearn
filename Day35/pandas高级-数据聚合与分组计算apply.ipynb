```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv('data/tips.csv')
tips.head(3)
```

```python
# 找出给定小费占总小费最多的
tips['tip_pct'] = tips['tip']/tips['total_bill']
tips.head(3)
```

```python
# 定义函数，筛选出小费占比最高的5条数据
def top(df,n=5,columns='tip_pct'):
    return df.sort_values(by=columns)[-n:]

top(tips)
```

```python
# 对数据按是否抽烟进行分组
grouped = tips.groupby('smoker')
# 通过apply对分组数据分别求占比前五的客户
grouped.apply(top)
```

```python
tips.groupby(['smoker','day'],group_keys=True).apply(top,columns='total_bill',n=2)
```

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
                    'data2': np.random.randn(1000)})
result = pd.cut(frame.data1,4)
result
```

```python
def get_stats(group):
    return {'min': group.min(),'max':group.max(),
            'count':group.count(), 'mean': group.mean()}
grouped = frame.data2.groupby(result)
grouped.apply(get_stats).unstack()
```
