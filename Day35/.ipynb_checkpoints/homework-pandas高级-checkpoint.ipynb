```python
### Step 3. 为了便由于分析，请获取到数据集中从列名为school到列名guardian之间的所有数据
# df.loc[:,['school':'guardian']]
df_sub = df.T.loc['school':'guardian'].T
df_sub
```

```python
# Step 4. 创建一个能实现字符串的首字母大写的lambda匿名函数
# f = lambda x:x[:1].upper()+x[1:].lower()
f = lambda x:x.capitalize()
```

```python
# Step 5. 将数据列 Mjob 和 Fjob中所有数据实现首字母大写
df[['Mjob','Fjob']].applymap(f)
```

```python
#Step 6. 经过第6步之后，为什么原来的dataframe数据中Mjob和Fjob列的数据仍然是小写的？简单说明原因，
# 并修改原始dataframe中的数据使得Mjob和Fjob列变为首字母大写
# 因为修改的是原始的视图复本，没有对原始数据进行修改
df[['Mjob','Fjob']] = df[['Mjob','Fjob']].applymap(f)
df
```

```python
# Step 7.创建一个名为majority函数，并根据age列数据返回一个布尔值添加到新的数据列，
# 列名为 legal_drinker (根据年龄这一列数据，大于17岁为合法饮酒)4
def majority(dataFrame):
    legal_drinker = dataFrame['age']>17
    dataFrame['legal_drinker'] = legal_drinker
```

```python
majority(df)
df
```
