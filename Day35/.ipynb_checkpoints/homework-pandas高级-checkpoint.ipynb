```python
# Step 4. åˆ›å»ºä¸€ä¸ªèƒ½å®žçŽ°å­—ç¬¦ä¸²çš„é¦–å­—æ¯å¤§å†™çš„lambdaåŒ¿åå‡½æ•°
# f = lambda x:x[:1].upper()+x[1:].lower()
f = lambda x:x.capitalize()
```

```python
#Step 6. ç»è¿‡ç¬¬6æ­¥ä¹‹åŽï¼Œä¸ºä»€ä¹ˆåŽŸæ¥çš„dataframeæ•°æ®ä¸­Mjobå’ŒFjobåˆ—çš„æ•°æ®ä»ç„¶æ˜¯å°å†™çš„ï¼Ÿç®€å•è¯´æ˜ŽåŽŸå› ï¼Œ
# å¹¶ä¿®æ”¹åŽŸå§‹dataframeä¸­çš„æ•°æ®ä½¿å¾—Mjobå’ŒFjobåˆ—å˜ä¸ºé¦–å­—æ¯å¤§å†™
# å› ä¸ºä¿®æ”¹çš„æ˜¯åŽŸå§‹çš„è§†å›¾å¤æœ¬ï¼Œæ²¡æœ‰å¯¹åŽŸå§‹æ•°æ®è¿›è¡Œä¿®æ”¹
df[['Mjob','Fjob']] = df[['Mjob','Fjob']].applymap(f)
df
```

```python
# Step 7.åˆ›å»ºä¸€ä¸ªåä¸ºmajorityå‡½æ•°ï¼Œå¹¶æ ¹æ®ageåˆ—æ•°æ®è¿”å›žä¸€ä¸ªå¸ƒå°”å€¼æ·»åŠ åˆ°æ–°çš„æ•°æ®åˆ—ï¼Œ
# åˆ—åä¸º legal_drinker (æ ¹æ®å¹´é¾„è¿™ä¸€åˆ—æ•°æ®ï¼Œå¤§äºŽ17å²ä¸ºåˆæ³•é¥®é…’)4
def majority(dataFrame):
    legal_drinker = dataFrame['age']>17
    dataFrame['legal_drinker'] = legal_drinker
```
