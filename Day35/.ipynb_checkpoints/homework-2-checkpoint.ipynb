### Step 1. 导入相应的模块

```python
from pandas import Series,DataFrame
import pandas as pd
```

### Step 3. 导入数据并赋值给变量 chipo，数据文件  dataset/chipotle.csv

```python
#  读取
chipo = pd.read_csv('datasets/chipotle.csv',sep='\t',error_bad_lines=False,skiprows=[1])
```

```python
chipo.head(3)
```

### Step 4. 计算出有多商品大于10美元

```python
fint = lambda x: float(x[1:])
chipo['priceint'] = chipo['item_price'].agg(fint)
chipo.head(3)
def count_p(df,columns='priceint'):
    cont = 0
    for i in list(df[columns]):
        if i>10:
            cont += 1
    return cont
count_p(chipo)

# chipo["item_price"] = chipo["item_price"].str.replace('#','')
# chipo["item_price"] = pd.to_numeric(chipo["item_price"]) 
# (chipo["item_price"]>10).value_counts()
```

### Step 6. 根据商品的价格对数据进行排序

```python
sortchipo = chipo.sort_values('priceint')
sortchipo.head(3)
```

### Step 7.在所有商品订单中 最贵商品的数量（quantity）是多少？

```python
chipo.sort_values('priceint')
quantity=list(sortchipo[-1:]['quantity'])[0]
quantity
```

### Step 8. 在所有的商品订购单中，数据商品Veggie Salad Bowl的订单数有几条？

```python
chipo.groupby('item_name').size()['Veggie Salad Bowl']
```

### Step 9. 在所有订单中，购买商品Canned Soda数量大于1的订单数有几条?

```python
(chipo['quantity'].groupby(chipo['item_name'])>10).values()
```
