```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np
from numpy import nan as NA
```

```python
data = DataFrame({
    "Qu1":[1,3,4,3,4],
    "Qu2":[2,3,1,2,3],
    "Qu3":[1,5,2,9,4]
})
```

```python
data
```

```python
#统计每一列数据中，每个数据出现的次数
result = data.apply(pd.value_counts).fillna(0)
result
```

```python
mask = data.apply(pd.value_counts).isnull()
```

```python
result[~mask] # 使用~进行取反
```

```python
data.apply(pd.value_counts).notnull()
```
