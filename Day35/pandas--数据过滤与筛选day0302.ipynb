```python
# np.random.seed(number) number 是生成随机数的种子
# np.random.randn() # 默认生成的随机种子数是当前时间的时间戳
# 定义一个种子数
np.random.seed(345673242) 
np.random.rand()
```

```python
np.random.seed(12345) # 定义种子数，保证数据一致性
data = DataFrame(np.random.randn(1000,4)) 
data
```

```python
# 数据简单的统计信息
data.describe()
```

```python
col = data[3]
col
```

```python
# 筛选出第四列中绝对值大于3的数
col[np.abs(col) > 3]
```

```python
# 查找出数据集中任意某一列的值出现了大于3的行数据
data[(np.abs(data)>3).any(1)] # 通过any() 筛选出每一行
```

```python
# 将所有的绝对值大于3的整数设置为3，绝对值大于3的复数。设置为-3
# 获取数据的符号
data[np.abs(data)>3] = np.sign(data)*3
data
```

```python
data[(np.abs(data) == 3).any(1)]
```
